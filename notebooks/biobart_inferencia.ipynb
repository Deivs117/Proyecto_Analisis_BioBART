# Inferencia BioBART (Demo de sustentación)\n\nNotebook base para demostrar el flujo Encoder-Decoder y generación biomédica paso a paso.

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM\n\nmodel_name = "GanjinZero/biobart-base"\ntokenizer = AutoTokenizer.from_pretrained(model_name)\nmodel = AutoModelForSeq2SeqLM.from_pretrained(model_name)
```

```python
texto = "Summarize: The patient was treated with broad-spectrum antibiotics and showed improved oxygen saturation."\ninputs = tokenizer(texto, return_tensors="pt", truncation=True)\n\nsalida = model.generate(\n    **inputs,\n    max_new_tokens=80,\n    do_sample=True,\n    temperature=0.8,\n    top_k=50,\n    top_p=0.95\n)\n\nprint(tokenizer.decode(salida[0], skip_special_tokens=True))
```
